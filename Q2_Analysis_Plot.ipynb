```python
import matplotlib.pyplot as plt
import numpy as np


baseline_ner = 0.98756
baseline_nli = 0.83718


sequential_results = {
    "NLI->NER": {"NLI": 0.353336, "NER": 0.991394},
    "NER->NLI": {"NLI": 0.829954, "NER": 0.0959046},
}


fixed_weights_results = {
    "Uniform": {"NLI": 0.8192562404482935, "NER": 0.9848839121867209},
    "0.25_0.75": {"NLI": 0.47, "NER": 0.87},
    "0.75_0.25": {"NLI": 0.86, "NER": 0.684},
}


dynamic_weights_results = {
    "Sigma_1": {"NLI": 0.788, "NER": 0.83},
    "Sigma_0.1": {"NLI": 0.45671, "NER": 0.90},
    "Sigma_0.5": {"NLI": 0.6286, "NER": 0.873449},
}


plt.figure(figsize=(14, 10))


plt.scatter(baseline_nli, baseline_ner, color='gray', marker='o', s=100, label='Baseline NER')
plt.scatter(baseline_ner, baseline_nli, color='orange', marker='o', s=100, label='Baseline NLI')


plt.scatter(sequential_results["NLI->NER"]["NLI"], sequential_results["NLI->NER"]["NER"], color='blue', marker='X', s=100, label='Sequential NLI->NER')
plt.scatter(sequential_results["NER->NLI"]["NER"], sequential_results["NER->NLI"]["NLI"], color='blue', marker='P', s=100, label='Sequential NER->NLI')


plt.scatter(fixed_weights_results["Uniform"]["NLI"], fixed_weights_results["Uniform"]["NER"], color='green', marker='s', s=100, label='Fixed Uniform')
plt.scatter(fixed_weights_results["0.25_0.75"]["NLI"], fixed_weights_results["0.25_0.75"]["NER"], color='green', marker='D', s=100, label='Fixed 0.25_0.75')
plt.scatter(fixed_weights_results["0.75_0.25"]["NLI"], fixed_weights_results["0.75_0.25"]["NER"], color='green', marker='^', s=100, label='Fixed 0.75_0.25')


plt.scatter(dynamic_weights_results["Sigma_1"]["NLI"], dynamic_weights_results["Sigma_1"]["NER"], color='red', marker='v', s=100, label='DWA Sigma=1')
plt.scatter(dynamic_weights_results["Sigma_0.1"]["NLI"], dynamic_weights_results["Sigma_0.1"]["NER"], color='red', marker='<', s=100, label='DWA Sigma=0.1')
plt.scatter(dynamic_weights_results["Sigma_0.5"]["NLI"], dynamic_weights_results["Sigma_0.5"]["NER"], color='red', marker='>', s=100, label='DWA Sigma=0.5')

x = np.linspace(*plt.xlim())
plt.plot(x, x, linestyle='--', color='black', linewidth=1, label='Reference Line')


plt.title('Comparison of Multitask Strategies on the NER and NLI Accuracies')
plt.xlabel('Accuracy on NLI Task')
plt.ylabel('Accuracy on NER Task')
plt.legend()


plt.tight_layout()
plt.grid()

plt.show()
```
